### Задание 2

```python
t = np.linspace(0,10,51)
f = np.cos(t)

plt.axis([0.5, 9.5, -2.5, 2.5])

plt.plot(t,f,color='green')
plt.title('График f(t)', fontsize=18, fontweight='bold')
plt.xlabel('Значения t', fontsize=14)
plt.ylabel('Значения f', fontsize=14)
plt.show()
```

### Задание 3

```python
x = np.linspace(-3,3,51)
y1 = x**2
y2 = np.array([elem * 2 + 0.5 for elem in x])
y3 = np.array([elem * (-3) - 1.5 for elem in x])
y4 = np.sin(x)
funcs = [y1, y2, y3, y4]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

for idx, ax in enumerate(ax.flatten()):
    ax.plot(x, funcs[idx])
    ax.set_title(f'График y{idx+1}', fontsize=12, fontweight='bold')
    ax.set_xlim([-5,5]) if idx == 0 else ax

fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

### Задание 4

```python
import pandas as pd
```

```python
df = pd.read_csv("creditcard.csv")  # https://www.kaggle.com/mlg-ulb/creditcardfraud
```

```python
plt.style.use('fivethirtyeight')
```

```python
df.Class.value_counts().plot(kind='bar')
plt.show()
```

```python
df.Class.value_counts().plot(kind='bar', logy=True)
plt.show()
```

```python
df.loc[df.Class == 0, "V1"].plot(kind='hist', ec='white', bins=20, alpha=0.5, logy=True, color='grey')
df.loc[df.Class == 1, "V1"].plot(kind='hist', ec='white', bins=20, alpha=0.5, logy=True, color='red')
plt.xlabel("Class")
plt.legend(["Class 0", "Class 1"])
plt.show()
```
